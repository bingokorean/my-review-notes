# Random Environment 

이번 실습에서는 OpenAI gym에서 제공하는 모든 환경 목록을 출력하고 다섯가지 환경에서 Random agent가 탐색하는 과정을 확인해 본다.

'wasd'로 환경을 변경 하실 수 있습니다.

# Import and predefine

```python
from IPython.display import clear_output
import gym

#list of envrionment
env_list = {
    '1':"Go9x9-v0",
    '2':"Go19x19-v0",
    '3':"Taxi-v2",
    '4':"FrozenLake-v0",
    '5':"FrozenLake8x8-v0",
    }

moveBindings = {
        'w':(1),
        'd':(1),
        's':(-1),
        'a':(-1),
    }
```

# Main function 

```python
envids = [spec.id for spec in gym.envs.registry.all()]
for envid in sorted(envids):
    print(envid)
print ""
print "There are a variety of environments provided by OpenAI gym"
print "you can choose several from this example"
print " 'w''d' next env, 's''a' prev env "
x = 0
while(1):
    key = raw_input() # input character
    clear_output() #clear output
    if key in moveBindings.keys():
        x = x + moveBindings[key]
        if x > 4 or x < 0:
            x = 0
    else:
        if (key == '\x03'):
            break
    #1 
    env = gym.make(env_list[env_list.keys()[x]]) # call environment 
    env.reset()
    for _ in range(50): # iteration time        
        env.render() #rendering method
        env.step(env.action_space.sample()) # take a random action
    print ""
    print "Environemt loaded"
    print env_list[env_list.keys()[x]]
    
```
