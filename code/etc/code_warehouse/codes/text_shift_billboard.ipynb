# Text-shift Billboard

* 텍스트는 1초당 왼쪽으로 한 칸씩 이동
* 첫 문구는 deco로 장식
* 입력된 시간 순간의 광고판을 출력
* 문구가 다 지나가면 반복

```python
def text_shift_ad(text, second):
    
    text_len = len(text)
    
    if int(second / text_len) % 2 == 0:
        text_off = False # 텍스트가 점점 등장하는 타임
    else:
        text_off = True # 텍스트가 점점 사라지는 타임
        
    deco = '_' * text_len # 데코레이션 (텍스트가 없는 부분)
    second = second % text_len # mod 기법을 활용한다 (홀짝 패턴만 알면)
    
    # 크게 두 가지 패턴을 가지기 때문에, 홀짝기반의 text_off 변수를 활용해서 다음과 같이 나눈다
    # 홀짝 패턴을 알고, mod를 통한 나머지를 통해 모든 가능한 경우의 수를 표현할 수 있다
    if text_off == False: # 텍스트가 점점 등장하는 타임    
        return deco[second:] + text[:second]   
    elif text_off == True: # 텍스트가 점점 사라지는 타임
        return text[second:] + deco[:second]
    
```

```python
# 2분동안 광고판을 표시해보아라.
for i in range(0, 120):
    print(text_shift_ad("Yeongsu, Fighting!!!", i))
```
